# Method : 1

```python
# Create function for items in the cart
def add_to_cart():
    items = {"Biscuits": {"Quantity": 5, "Cost_per_item": 2.50},                     # Create a dictionary.
             "Cereals": {"Quantity": 10, "Cost_per_item": 3.50},
             "Chicken": {"Quantity": 20, "Cost_per_item": 13.30},
             "Oats": {"Quantity": 20, "Cost_per_item": 3.90},
             "Rice":{"Quantity": 20, "Cost_per_item": 4.50}}


    print("%50s" %("Essential Items Available"))                                    # Display available items
    print("________________________________________________________________________________________________")
    print("%20s" "%20s" "%30s" "%20s" %("S.No","Items","Quantity","Cost_per_Item"))
    print("__________________________________________________________________________________________________")
    print("%20s" "%30s" "%15s" "%25s" %("1", "Biscuits: a pack of 20", items["Biscuits"]["Quantity"],"2.50"))
    print("%20s" "%30s" "%15s" "%25s" %("2", "Cereals: 2 pounds/pack", items["Cereals"]["Quantity"],"3.50"))
    print("%20s" "%30s" "%15s" "%25s" %("3", "Chicken: 1 pounds/pack", items["Chicken"]["Quantity"],"13.30"))
    print("%20s" "%30s" "%15s" "%25s" %("4", "Oats: 4 pounds/pack", items["Oats"]["Quantity"],"3.90"))
    print("%20s" "%30s" "%15s" "%25s" %("5", "Rice: 2 pounds/pack", items["Rice"]["Quantity"],"4.50"))
             
    print("\n")
    Biscuit_Qty = int(input("How many packets of biscuits are required ? : "))       # User Input for quantity required
    Cereals_Qty = int(input("How many packets of cereals are required ? : "))
    Chicken_Qty = int(input("How many packets of frozen chicken are required ? : "))         
    Oats_Qty = int(input("How many packets of oats are required ? : "))         
    Rice_Qty = int(input("How many packets of rice are required ? : "))
             
    if Biscuit_Qty > items["Biscuits"]["Quantity"] or Cereals_Qty > items["Cereals"]["Quantity"] or Chicken_Qty > items["Chicken"]["Quantity"] or Oats_Qty > items["Oats"]["Quantity"] or Rice_Qty > items["Rice"]["Quantity"]:
             print("Items are out of stock")                                         # Stock Availability Check
             return
    
    print("\nFill in your delivery details")                                         # User Input for Delivery Details
    Name = input("Enter your name : ")
    Address = input("Enter your address : ")
    Distance = input("Enter Distance from centre 0-5 / 5-10 / 10-15 / Above 15 :")
    
    Biscuit_Cost = Biscuit_Qty * 2.50                                                # Cost of each item = Quantity X Price
    Cereals_Cost = Cereals_Qty * 3.50 
    Chicken_Cost = Chicken_Qty * 13.30          
    Oats_Cost = Oats_Qty * 3.90
    Rice_Cost = Rice_Qty * 4.50
    Total_Products_Cost = Biscuit_Cost + Cereals_Cost + Chicken_Cost + Oats_Cost + Rice_Cost
    Total_Cost = Total_Products_Cost                                                 # Total cost of all products
    
    print("\n")
    print("%50s" %("Bill"))                                                          # Display Bill
    print("_______________________________________________________________________________________________________")
    print("%20s" "%20s" "%20s" "%20s" %("S.No", "Items", "Quantity", "Total Cost"))
    print("_______________________________________________________________________________________________________")
    print("%20s" "%20s" "%20s" "%20s" "%2s"  %("1", "Biscuits", Biscuit_Qty, "$", Biscuit_Cost))
    print("%20s" "%20s" "%20s" "%20s" "%2s"  %("2", "Cereals", Cereals_Qty, "$", Cereals_Cost))
    print("%20s" "%20s" "%20s" "%20s" "%2s"  %("3", "Chicken", Chicken_Qty, "$", Chicken_Cost))
    print("%20s" "%20s" "%20s" "%20s" "%2s"  %("4", "Oats", Oats_Qty, "$", Oats_Cost))
    print("%20s" "%20s" "%20s" "%20s" "%2s"  %("5", "Rice", Rice_Qty, "$", Rice_Cost))
    
    print("\n")
    print("Total Cost = $", Total_Cost)
    
    Distance_0_to_5 = 10
    Distance_5_to_10 = 12
    Distance_10_to_15 = 14
    Distance_Above_15 = 20
    
    # Calculate total bill amount by adding total cost and delivery charges.
    if Distance == "0-5":
        print("Total Bill Amount = Total Cost + Delivery Charges = ", Total_Cost + Distance_0_to_5)
    elif Distance == "5-10":
        print("Total Bill Amount = Total Cost + Delivery Charges = ", Total_Cost + Distance_5_to_10)
    elif Distance == "10-15":
        print("Total Bill Amount = Total Cost + Delivery Charges = ", Total_Cost + Distance_10_to_15)
    elif Distance == "Above 15":
        print("Total Bill Amount = Total Cost + Delivery Charges = ", Total_Cost + Distance_Above_15)
    else:
        print("Products are not deliverable.The distance is beyond the allowed delivery distance")
    
    # Print Delivery Details 
    print("\n")
    print("Name : ",Name)                                                             
    print("Address : ",Address)
    
    # Using while loop to check if customer wants to continue shopping or not. If not, break the loop.
    while True:                                                                       
        finish = input("\nDo you wish to continue shopping ? 'yes/no': ")
        if finish == 'yes':
            continue
        else:
            print("\nThankyou for shopping. Have a nice day.")
            break
            
    
# Call the function    
add_to_cart()                
             
```

```python
# Create a dictionary with item name as keys and value as tuple of quantity and price per item.
items_av = {"Biscuits": (5, 2.5), "Cereals": (10, 3.50), "Chicken": (20, 13.30), "Oats": (20, 3.90), "Rice": (30, 4.50)}

# Create inventory function which displays the items available for purchase along with their quantities and prices.
# It iterates over the items_av dictionary and prints each item's details.

def inventory():                         
    print("%40s" % ("Essential Items Available"))  
    print("___________________________________________________________________")
    print("%10s %15s %12s %15s" % ("S.no", "Items", "Quantity", "Cost_per_item"))
    print("___________________________________________________________________")
    print()

    for i, (items, (quantity, price)) in enumerate(items_av.items(), start=1):
        print("%10s %15s %10s %10s" % (i, items, quantity, price))

inventory()      # Call inventory function

# Create delivery details function which accpet user input details regarding delivery.
def delivery_details():
    print("\nFill in your delivery details")  
    Name = input("\nEnter your name : ")
    Address = input("\nEnter your address : ")
    Distance = input("\nEnter Distance from centre 0-5 / 5-10 / 10-15 / Above 15 :")
    return Name, Address, Distance

Name, Address, Distance = delivery_details()

# Create bill function with two parameters.This function calculates and prints the bill for the items purchased.
def bill(cart, Distance):
    total_cost = 0
    print("%50s" % ("Bill"))
    print("_______________________________________________________________________________________")
    print("%20s %20s %20s %20s" % ("S.No", "Items", "Quantity", "Total Cost"))
    print("_______________________________________________________________________________________")

    for i, (item, (quantity, price)) in enumerate(cart.items(), start=1):
        total_item_cost = quantity * price
        total_cost += total_item_cost
        print("%20s %20s %20s %20s" % (i, item, quantity, total_item_cost))

 # Conditions for delivery charge       
    delivery_charge = 0
    if Distance == "0-5":
        delivery_charge = 10
    elif Distance == "5-10":
        delivery_charge = 12
    elif Distance == "10-15":
        delivery_charge = 14
    elif Distance == "Above 15":
        delivery_charge = 20
    else:
        print("Products are not deliverable. The distance is beyond the allowed delivery distance")
        return

    total_bill_amount = total_cost + delivery_charge
    print("\nTotal Cost: $", total_cost)

    print("\nDelivery Charge: $", delivery_charge)
    
    print("\nTotal Bill Amount = Total Cost + Delivery Charges = $", total_bill_amount)

    print("\nName : ", Name)
    print("Address : ", Address)

# Create function add_to_cart which allows the user to add items to their shopping cart.
def add_to_cart1():
    cart = {}
    while True:
        item = input("\nEnter the name of item: ")
        if item not in items_av:
            print("Item not available in inventory")
            continue
        quantity = int(input("Enter the quantity required: "))
        cart[item] = (quantity, items_av[item][1]) 
        finish = input("Do you wish to continue shopping? 'yes/no': ")
        if finish == 'no':
            print("\nThank you for shopping. Have a nice day.")
            break
    bill(cart, Distance)
    
    while True:                                                                       
        finish = input("\nDo you wish to continue shopping ? 'yes/no': ")
        if finish == 'yes':
            continue
        else:
            print("\nThankyou for shopping. Have a nice day.")
            break
add_to_cart1()
```
